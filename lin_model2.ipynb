```python
# Standard imports for data analysis packages in Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This enables inline Plots
%matplotlib inline

# Limit rows displayed in notebook
pd.set_option('display.max_rows', 10)
pd.set_option('display.precision', 2)
```

```python
df = pd.read_csv('saleprice_df.csv')
```

```python
# First, without dates or neighborhoods
cat_var = ['INSTRUMENT', 'CLASS', 'D_CLASS', 'OFCARD', 'TAX_DIST', 'STORY']

df_cat = df[cat_var]
df_cat['D_CLASS_s'] = df_cat['D_CLASS'].astype('str')
df_cat['OFCARD_s'] = df_cat['OFCARD'].astype('str')
df_cat['STORY_s'] = df_cat['STORY'].astype('str')

df_catd = pd.get_dummies(df_cat, dummy_na=True)

df_cat.drop(['D_CLASS', 'OFCARD', 'STORY'], axis=1)

df_d = pd.concat([df, df_catd], axis=1)
df_d1 = df_d.drop(df[cat_var], axis=1)
```

```python
df_d2 = df_d1.drop(['SITE_NBR', 'SITE_DIR', 'SITE_NAME', 'SITE_MODE', 'SITE_MORE','NBHD_1_x'], axis=1)
df_d2['RECENT_RENO'] = np.where(df_d2['CCAGE_RM'] >= 2000, 1, 0)
df_d2['HOME_AGE'] = 2015 - df_d2.CCYRBLT
df_d2['SALES_YR'] = df_d2['RECEPTION_DATE'].map(lambda x : str(x)[:4]).astype('str')
df_d2['SALES_MONTH'] = df_d2['RECEPTION_DATE'].map(lambda x : str(x)[4:6]).astype('str')

df_d2a = df_d2[df_d2.SALES_YR.isin(['2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'])]

df_d3 = df_d2a.drop(['CCAGE_RM', 'RECEPTION_DATE', 'RECEPTION_NUM', 'SCHEDNUM', 'Unnamed: 0', 'CCYRBLT'], axis=1)
```

```python
col_dist = {}
def get_col_dist(col_name):
    excl_null_mask = df_d3[col_name].notnull()
    row_count = df_d3[excl_null_mask][col_name].size
    col_data = {}
    col_data['prob'] = (df_d3[excl_null_mask][col_name].value_counts() / row_count).values
    col_data['values'] = (df_d3[excl_null_mask][col_name].value_counts() / row_count).index.values
    return col_data


col_dist['FULL_B'] = get_col_dist('FULL_B')
col_dist['HLF_B'] = get_col_dist('HLF_B')
col_dist['HOME_AGE'] = get_col_dist('HOME_AGE')
```

```python
def impute_cols(val, options):
    if np.isnan(val):
        return np.random.choice(options['values'], p=options['prob'])
    return val

def impute_FULL_B(val):
    return impute_cols(val, col_dist['FULL_B'])

def impute_HLF_B(val):
    return impute_cols(val, col_dist['HLF_B'])

def impute_HOME_AGE(val):
    return impute_cols(val, col_dist['HOME_AGE'])
```

```python
df_d3['FULL_B_imp'] = df_d3.FULL_B.map(impute_FULL_B)
df_d3['HLF_B_imp'] = df_d3.HLF_B.map(impute_HLF_B)
df_d3['HOME_AGE_imp'] = df_d3.HOME_AGE.map(impute_HOME_AGE)
```

```python
def impute_numeric_cols(col_data, col_name):
    na_row_count = col_data.isnull().sum()
    impute_vals = np.random.normal(col_data.mean(), col_data.std(), na_row_count)
    return impute_vals

LAND_SQFT_rows_mask = df_d3['LAND_SQFT'].isnull()
df_d3.loc[LAND_SQFT_rows_mask, 'LAND_SQFT'] = impute_numeric_cols(df_d3['LAND_SQFT'], 'LAND_SQFT')

df_d4 = df_d3.drop(['FULL_B','HLF_B','HOME_AGE'], axis=1)
```

```python
held_vars = ['NBHD_1_CN_x', 'SALES_YR', 'SALES_MONTH']
df_held = df_d4[held_vars]

df_d5 = df_d4.drop(df_d4[held_vars], axis=1)
```

```python
# Import regressor
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
```

```python
X = df_d5.drop(['SALE_PRICE'], axis=1)
y = df_d5['SALE_PRICE']

X_train1, X_test1, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=36)
```

```python
linest1 = LinearRegression()

linest1.fit(X_train1, y_train)

linest1.score(X_test1, y_test)
```

```python
y_pred1 = linest1.predict(X_test1)

fig, ax = plt.subplots(1, 1)

ax.set_title('Without year, month, NBHD')

ax.scatter(y_pred1, y_test, alpha=0.7)
ax.set_xlabel('Predicated sale price')
ax.set_ylabel('Actual sale price')

ax.set_xscale('log')
ax.set_yscale('log')
```
