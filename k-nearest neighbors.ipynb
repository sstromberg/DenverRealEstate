```python
# Standard imports for data analysis packages in Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This enables inline Plots
%matplotlib inline

# Limit rows displayed in notebook
pd.set_option('display.max_rows', 10)
pd.set_option('display.precision', 2)
```

```python
df = pd.read_csv('saleprice_df.csv')
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import cross_val_score
```

```python
df.info()
```

```python
df_k = df.drop(['SITE_NBR', 'SITE_DIR', 'SITE_NAME', 'SITE_MODE', 'SITE_MORE','NBHD_1_x', 'RECEPTION_NUM', 
                    'SCHEDNUM', 'Unnamed: 0'], axis=1)
```

```python
df_k.info()
```

```python
df_k['SALES_YR'] = df_k['RECEPTION_DATE'].map(lambda x : str(x)[:4]).astype('int')
df_k['SALES_MONTH'] = df_k['RECEPTION_DATE'].map(lambda x : str(x)[4:6]).astype('int')
df_k['RECENT_RENO'] = np.where(df_k['CCAGE_RM'] >= 2000, 1, 0)
df_k['HOME_AGE'] = 2015 - df_k.CCYRBLT

df_k2 = df_k.drop(['RECEPTION_DATE', 'CCAGE_RM', 'CCYRBLT'], axis=1)

df_k2.info()
```

```python
df_k3 = df_k2.drop(['INSTRUMENT', 'CLASS', 'D_CLASS', 'NBHD_1_CN_x', 'OFCARD', 'TAX_DIST'], axis=1)
```

```python
col_dist = {}
def get_col_dist(col_name):
    excl_null_mask = df_k3[col_name].notnull()
    row_count = df_k3[excl_null_mask][col_name].size
    col_data = {}
    col_data['prob'] = (df_k3[excl_null_mask][col_name].value_counts() / row_count).values
    col_data['values'] = (df_k3[excl_null_mask][col_name].value_counts() / row_count).index.values
    return col_data


col_dist['FULL_B'] = get_col_dist('FULL_B')
col_dist['HLF_B'] = get_col_dist('HLF_B')
col_dist['HOME_AGE'] = get_col_dist('HOME_AGE')


def impute_cols(val, options):
    if np.isnan(val):
        return np.random.choice(options['values'], p=options['prob'])
    return val

def impute_FULL_B(val):
    return impute_cols(val, col_dist['FULL_B'])

def impute_HLF_B(val):
    return impute_cols(val, col_dist['HLF_B'])

def impute_HOME_AGE(val):
    return impute_cols(val, col_dist['HOME_AGE'])

df_k3['FULL_B_imp'] = df_k3.FULL_B.map(impute_FULL_B)
df_k3['HLF_B_imp'] = df_k3.HLF_B.map(impute_HLF_B)
df_k3['HOME_AGE_imp'] = df_k3.HOME_AGE.map(impute_HOME_AGE)
```

```python
def impute_numeric_cols(col_data, col_name):
    na_row_count = col_data.isnull().sum()
    impute_vals = np.random.normal(col_data.mean(), col_data.std(), na_row_count)
    return impute_vals

LAND_SQFT_rows_mask = df_k3['LAND_SQFT'].isnull()
df_k3.loc[LAND_SQFT_rows_mask, 'LAND_SQFT'] = impute_numeric_cols(df_k3['LAND_SQFT'], 'LAND_SQFT')

df_k4 = df_k3.drop(['FULL_B','HLF_B','HOME_AGE'], axis=1)
```

```python
X = df_k4.drop(['SALE_PRICE'], axis=1)
y = df_k4['SALE_PRICE']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=30)
```

```python
knn_mod = KNeighborsRegressor(n_neighbors=6, weights='distance')
```

```python
%%time
knn_mod.fit(X_train, y_train)
```

```python
knn_mod.score(X_test, y_test)
```

```python
%%time
cross_val_score(knn_mod, X_train, y_train)
```

```python
y_pred = knn_mod.predict(X_test)
```

```python
x = range(100000,5000000,10000)

fig, ax = plt.subplots(1, 1, figsize=(6, 4))

ax.plot(x, x, c='r')

ax.scatter(y_pred, y_test, c='purple', alpha=0.2)
ax.set_xlabel('Predicted sale price')
ax.set_ylabel('Actual sale price')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlim([50000, 10000000])
ax.set_ylim([50000, 10000000])

ax.set_title('KNN Predictions, 6 Neighbors, Distance-Weighted')
```
